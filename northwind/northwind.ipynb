```python
# reading all excel sheets into different dataframes 
df_customers=pd.read_excel("Northwind.xlsx",sheet_name=1)
df_orders=pd.read_excel("Northwind.xlsx",sheet_name=4)
df_Categories=pd.read_excel("Northwind.xlsx",sheet_name=0)
df_Employees=pd.read_excel("Northwind.xlsx",sheet_name=2)
df_orderdetails=pd.read_excel("Northwind.xlsx",sheet_name=3)
df_products=pd.read_excel("Northwind.xlsx",sheet_name=5)
```

```python

import warnings 
warnings.filterwarnings("ignore")

#  maybe we need this column in some analysis so just replacing the missing values with unkown 
df_customers['region'].fillna("Unknown", inplace=True)
# and the fax column is not used much but just keeping the column as it could be used in analysis if necessary 
df_customers['fax'].fillna('Not Provided', inplace=True)
df_customers["postalcode"].fillna('Not Provided',inplace=True)
```

```python
df_customers.columns = df_customers.columns.str.lower()
df_customers.rename(columns={'contactname': 'contact_name', 'companyname': 'company_name'}, inplace=True)
df_customers.rename(columns={"customerid": "CustomerID"}, inplace=True)

```
