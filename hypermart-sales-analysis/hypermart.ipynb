 Products Having the Highest Sales but Low or Negative Profit Margins

```python

columns_2 = df.groupby('Product Name')[['Sales', 'Profit']].sum().reset_index()
low_profit = columns_2[columns_2['Profit'] <= 0]
top_low_profit = low_profit.sort_values(by='Sales', ascending=False)
print(top_low_profit.head(10))

plt.figure(figsize=(12, 6))
sns.barplot(data=top_low_profit.head(10), x='Sales', y='Product Name', hue='Product Name', palette='Reds_r', legend=False)
plt.title('Top 10 Products with High Sales but Low/Negative Profit')
plt.xlabel('Total Sales')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

**Insight:** That indicates that most sales are of lower value, and a few big items contribute to revenue. There are several products that, despite generating high sales, have low or negative profit margins. This shows potential pricing or cost issues with these specific items, and they may require re-evaluation of pricing strategy or cost control measures to ensure profitability.

Does Shipping Mode Impact Order Delivery Time

```python
#converting from string form to dateform 
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=False, errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=False, errors='coerce')

df['Delivery Time'] = (df['Ship Date'] - df['Order Date']).dt.days
shipping = df.groupby('Ship Mode')['Delivery Time'].mean().reset_index()
shipping =shipping.sort_values(by='Delivery Time', ascending=False)
print(shipping)

# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(data=shipping, x='Delivery Time', y='Ship Mode', hue='Ship Mode', palette='Blues', legend=False)
plt.title('Average Delivery Time by Shipping Mode')
plt.xlabel('Average Delivery Time (Days)')
plt.ylabel('Shipping Mode')
plt.tight_layout()
plt.show()
```

**Insight:**
The bar plot clearly shows that Standard Class has the longest average delivery time, followed by Second Class, First Class, and Same Day.
This indicates that Shipping Mode significantly impacts delivery time, with faster modes like Same Day and First Class offering quicker delivery, while Standard Class takes the longest.

 Monthly Sales Trend (Seasonal Patterns) 

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed', errors='coerce')
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month_name()
heatmap_data = df.pivot_table(values='Sales', index='Month', columns='Year', aggfunc='sum')
from calendar import month_name
ordered_months = list(month_name)[1:]  # month_name[0] is empty string
heatmap_data = heatmap_data.reindex(ordered_months)

# Plotting 
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.0f', linewidths=.5)
plt.title('Monthly Sales Trend (Heatmap)')
plt.xlabel('Year')
plt.ylabel('Month')
plt.tight_layout()
plt.show()
```

**Insight:**
The heatmap reveals seasonal sales patterns, with consistent peaks observed in the months of November and December across multiple years.
This indicates a strong year-end buying trend, possibly due to holiday seasons or promotional events

 Sales Distribution 

```python
# Calculate 95th percentile to remove outliers
threshold = df['Sales'].quantile(0.95)
filtered_df = df[df['Sales'] <= threshold]

# Plot after removing outliers
plt.figure(figsize=(8, 5))
sns.histplot(filtered_df['Sales'], bins=40, kde=True, color='skyblue')
plt.title('Distribution of Sales (Without Outliers)')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

**Insight:**
The sales distribution (after removing the top 5% outliers)  it indicates that most transactions have lower sales values, while high-value sales are relatively rare.

 Which Customer Segment generates the highest Sales and Profit?

```python
segment_sales_profit = df.groupby('Segment')[['Sales', 'Profit']].sum().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
segment_sales_profit_melted = segment_sales_profit.melt(id_vars='Segment', value_vars=['Sales', 'Profit'])
sns.barplot(data=segment_sales_profit_melted, x='Segment', y='value', hue='variable', palette='Set2')
plt.title('Total Sales and Profit by Customer Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```

**Insight:**
The bar plot shows that the Consumer segment generates the highest total Sales, while the Corporate segment shows the highest Profit. This indicates that although consumers buy more, corporate clients contribute more efficiently to profitability.


-------------------------------------------------------------------------------------------------------------------------------------------------------

What is the repeat purchase behaviour of customers in each segment?

```python
repeat_customers = df.groupby(['Segment', 'Customer ID'])['Order ID'].nunique().reset_index()
avg_repeat_orders = repeat_customers.groupby('Segment')['Order ID'].mean().reset_index()

# Plotting
plt.figure(figsize=(8, 5))
sns.barplot(data=avg_repeat_orders, x='Segment', y='Order ID', hue='Segment', palette='coolwarm', legend=False)
plt.title('Average Number of Orders per Customer in Each Segment')
plt.xlabel('Customer Segment')
plt.ylabel('Average Repeat Orders')
plt.tight_layout()
plt.show()
```

**Insight:**
The bar plot reveals that Corporate and Home Office customers have slightly higher average repeat purchases compared to Consumers.

Are there customer segments in underperforming regions needing targeted marketing?

```python
region_segment = df.groupby(['Region', 'Segment'])[['Sales', 'Profit']].sum().reset_index()

# Use keyword arguments for pivot
pivot_data = region_segment.pivot(index='Region', columns='Segment', values='Profit')

# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_data, cmap='RdYlGn', annot=True, fmt=".0f")
plt.title('Profit by Segment and Region')
plt.xlabel('Customer Segment')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

**Insight:**
The heatmap reveals that Corporate and Home Office segments in the Central region show  low or negative profits, so these are the targets for improved marketing strategies. whereas, Consumer segments in the West and East regions are performing well, which tells us that existing strategies are effective there.
